What do we mean by having clean/Tidy Data?
1. One visual tell-tale sign is that the dataset is structured into distinct input variables and response variables.
Note:Take into account the context of your dataset



# **Common Problems with Messy Datasets**



1. Column headers are values, not variable names
2. Multiple variables are stored in one column
3. Variables are stored in both rows and columns
4. Mulitple types of observational units are stored in the  same table
5. A single observational unit is stored in multiple tables.
